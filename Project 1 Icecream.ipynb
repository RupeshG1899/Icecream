```python
icecream = {'Cone Type': {'Plain Cone' : 1.5, 'Waffle Cone' : 2, 'Cup' : 1}, 
            'Scoop Flavours' : {'Vanilla' : 0.5, 'Strawberry' : 0.5, 'Chocolate' : 0.5, 'Caramel' : 0.5, 'Mint' : 0.5, 'Rainbow' : 0.5, 'Coffee' : 0.5, 'Bubble gum' : 0.5}, 
            'Topping Choices' :{'Peanuts' : 0.75, 'Caramel Sauce' : 0.5, 'Rainbow Sprinkles' : 0.5, 'Pecan' : 1, 'Chocolate Sprinkles' : 0.5}}
```

```python
Name_of_the_customer = input('Enter the name of customer: ')
dict1 = {}
order = ''
price = 0
dict2 = {}
cone_type = ''
while True: 
    cone_type = input('Enter Cone Type or no: ')
    if cone_type != 'no':
        try:
            if cone_type in icecream['Cone Type']:
                order = cone_type
                price = icecream['Cone Type'][order]
                dict2[order] = price
                if order in dict1:
                    dict1[order] += 1
                else:
                    dict1[order] = 1 
                j = 0
                while j != 3:
                    scoop_flavour = input('Enter Scoop Flavour (max 3) or no: ')
                    if scoop_flavour != 'no':
                        try:
                            if scoop_flavour in icecream['Scoop Flavours']:
                                order = scoop_flavour
                                price = icecream['Scoop Flavours'][order]
                                dict2[order] = price
                                if order in dict1:
                                    dict1[order] += 1
                                else:
                                    dict1[order] = 1
                                j += 1
                            else:
                                raise ValueError('Invalid Scoop Flavour') 
                        except ValueError as e:
                            print(e)
                    else:
                        break
                k = 0
                while k != 4:
                    type_of_topping = input('Enter the Topping Choice (max 4) or no: ')
                    if type_of_topping != 'no':
                        try:
                            if type_of_topping in icecream['Topping Choices']:
                                order = type_of_topping
                                price = icecream['Topping Choices'][order]
                                dict2[order] = price
                                if order in dict1:
                                    dict1[order] += 1
                                else:
                                    dict1[order] = 1
                                k += 1
                            else:
                                raise ValueError('Invalid Topping Choice')
                        except ValueError as e:
                            print(e)
                    else:
                        break
            else:
                raise ValueError('Invalid Cone Type')
        except ValueError as e:
            print(e)
    else:
        break
print(dict1)
lst = []
print('Below Zero - ice cream store'.center(38))
print('Receipt'.center(38))
print("Item's ordered       | Quantity | Price")
print('=======================================')
for order, quantity in dict1.items():
    price = quantity * dict2[order]
    lst.append(price)
    print(f'{order:20} | {quantity:<8} : {price}')
print('=======================================')
print('Total price'.ljust(30),':', sum(lst))
print('Thank You'.center(38))
```
